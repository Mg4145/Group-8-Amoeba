```python
def get_goal_reached(df):
    
    true = 0
    df_length = len(df["Goal Reached"])
    
    for val in df["Goal Reached"]:
        if val == True:
            true += 1
    true /= df_length
    
    return round(true * 100, 2)
```

```python
# Get percentages
g1_gr = get_goal_reached(g1)
g2_gr = get_goal_reached(g2)
g3_gr = get_goal_reached(g3)
g4_gr = get_goal_reached(g4)
g5_gr = get_goal_reached(g5)
g6_gr = get_goal_reached(g6)
g7_gr = get_goal_reached(g7)
g8_gr = get_goal_reached(g8)
g9_gr = get_goal_reached(g9)


# Save them to a list
goal_reach = []
goal_reach.append(g1_gr)
goal_reach.append(g2_gr)
goal_reach.append(g3_gr)
goal_reach.append(g4_gr)
goal_reach.append(g5_gr)
goal_reach.append(g6_gr)
goal_reach.append(g7_gr)
goal_reach.append(g8_gr)
goal_reach.append(g9_gr)
```

```python
# Plot Overall Success for all Groups

# goal_reach
groups = list(range(1, 10))

plt.figure(figsize=(10, 5))
plt.xlabel("Groups")
plt.ylabel("Percentage")
plt.title("Groups Percentage of Success Overall")

plt.bar(groups, 
        goal_reach, 
        color = ['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'maroon'])

plt.xticks(groups)

plt.show()
```

```python
def get_timeout(df):
    df_len = len(df["Timeout"])
    timeout = 0
    for val in df["Timeout"]:
        if val == True:
            timeout += 1
    timeout /= df_len
    
    return round(timeout * 100, 2) 
```

```python
def get_timeout_size(df, size):
    new_df = df.loc[df["Starting Size"] == size, ["Timeout"]]
    
    true = 0
    df_length = len(new_df["Timeout"])
    
    for val in new_df["Timeout"]:
        if val == True:
            true += 1
    true /= df_length
    return round(true * 100, 2)
```
